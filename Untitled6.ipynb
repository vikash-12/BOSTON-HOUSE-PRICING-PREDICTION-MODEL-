# Load the Boston House Pricing Dataset

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
df = pd.DataFrame(boston.data, columns = boston.feature_names)
```

```python
df
```

```python
df['PRICE'] = boston.target
```

```python
sns.regplot(x='PTRATIO',y='PRICE',data = df)
```

```python
## DEPENDANT AND INDEPANDANT FEATURES
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
## train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state = 42)
```

```python
x_train
```

```python
x_test
```

```python
## standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
x_test
```

MODEL TRAIN

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
## print coefficient and intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
##prediction with test data
reg_pred = regression.predict(x_test)
```

```python
reg_pred
```

```python
## plot scatter plot between test and predicted
plt.scatter(y_test,reg_pred)
```

```python
##residuals
residuals = y_test - reg_pred
```

```python
sns.displot(residuals, kind='kde')
```

```python
## scatter plot betwwen reg_pred and residuals
plt.scatter(reg_pred, residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(reg_pred,y_test))
print(mean_squared_error(reg_pred,y_test))
```
